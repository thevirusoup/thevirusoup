```python
# Function to check if sequence headers match between two files
def compare_fasta_headers(file1, file2):
    headers1 = [line.strip() for line in open(file1) if line.startswith(">")]
    headers2 = [line.strip() for line in open(file2) if line.startswith(">")]

    if headers1 == headers2:
        print("✅ Headers match, no misplaced sequences.")
    else:
        print("❌ Headers are mismatched. Sequences may be displaced.")

# Compare toxic and random FASTA files
compare_fasta_headers("toxic_DNA.fasta", "random_DNA.fasta")
compare_fasta_headers("toxic_protein.fasta", "random_Protien.fasta")
```

```python
# Function to read a FASTA file and store sequences in a dictionary
def read_fasta(file_path):
    sequences = {}
    with open(file_path, "r") as file:
        seq_id = None
        seq_data = []
        for line in file:
            line = line.strip()
            if line.startswith(">"):
                if seq_id:
                    sequences[seq_id] = "".join(seq_data)
                seq_id = line[1:]  # Remove '>'
                seq_data = []
            else:
                seq_data.append(line)
        if seq_id:
            sequences[seq_id] = "".join(seq_data)
    return sequences

# Function to reorder sequences based on reference file
def reorder_sequences(reference_file, to_correct_file, output_file):
    reference_sequences = read_fasta(reference_file)  # Read correct file
    to_correct_sequences = read_fasta(to_correct_file)  # Read incorrect file

    corrected_sequences = []

    for seq_id in reference_sequences:
        if seq_id in to_correct_sequences:
            corrected_sequences.append(f">{seq_id}\n{to_correct_sequences[seq_id]}")
        else:
            print(f"⚠ Warning: {seq_id} is missing in the incorrect file!")

    with open(output_file, "w") as file:
        file.write("\n".join(corrected_sequences) + "\n")

    print(f"✅ Corrected file saved as: {output_file}")

# Correct the random files based on toxic files
reorder_sequences("toxic_DNA.fasta", "random_DNA.fasta", "corrected_random_DNA.fasta")
reorder_sequences("toxic_protein.fasta", "random_Protien.fasta", "corrected_random_Protien.fasta")
```
